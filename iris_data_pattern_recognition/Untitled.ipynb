## 3. Univariate Plots - To understand each plot better

### 3.1 Box Plot - Distribution of input attributes through quartiles and find out outliers

```python
# box plot and whisker plots give idea about distribution of input attributes
# box plot using pandas
df.plot(kind='box', subplots=True, layout = (2,2), sharex=False, sharey=False)
```

```python
plot_data = df[['sepal_length','sepal_width', 'petal_length','petal_width']]
```

```python
# box plot using matplotlib
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
bplot = ax.boxplot(plot_data, notch=True, patch_artist=True, vert=False, labels=('sepal_length','sepal_width', 'petal_length', 'petal_width'), autorange=True)
colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']
for patch , color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

    
# changing color and linewidth of whiskers
for whisker in bplot['whiskers']:
    whisker.set(color ='#8B008B',
                linewidth = 1.5,
                linestyle =":")


# changing color and linewidth of caps
for cap in bplot['caps']:
    cap.set(color ='#8B008B',
            linewidth = 2)
    

# changing color and linewidth of
# medians
for median in bplot['medians']:
    median.set(color ='red',
               linewidth = 3)
  

# changing style of fliers
for flier in bplot['fliers']:
    flier.set(marker ='D',
              color ='#e7298a',
              alpha = 0.5)
    

# Adding title
plt.title("Box Plot")

    
    
    
    
plt.show(bplot)
```

```python
# Box Plot using seaborn 
sns.set_style('whitegrid')
sns.boxplot(data=plot_data, orient='h')
```

```python
fig_px = px.box(plot_data, points="all", orientation='v', width=900, height=500)
fig_px.show()
# https://plotly.github.io/plotly.py-docs/generated/plotly.express.box.html
```

> In the box plot we can see that there are outliers in the sepal width the values go past upper fence and lower fence

### 3.2 Histogram - Distribution of input attribute through their bin, we find if the distribution of atttribute follows Gaussian or other distribution

A histogram might not be useful in all cases
Histogram, counts, distplot
1. Histplot - A histplot is used when the plot variable is continuous. 
2. Count - A count is used when the variables are discrete
3. Distplot - Also used for distplot

```python
# histograms using matplotlib
df.hist()
hsg = plt.gcf() # gets current figure
hsg.set_size_inches(12,6)
plt.show()

# below we can see that sepal length and sepal width have approximately Gaussian distribution 
```

```python
# histograms using matplotlib



fig = plt.figure(figsize=(20, 10))
for i in range(len(plot_data.columns)):
    ax = fig.add_subplot(2,2,i+1)
    ax.hist(
        plot_data[plot_data.columns[i]],
        density=True, 
        color=None
    )
    ax.set_title(plot_data.columns[i])
plt.show()

# below we can see that sepal length and sepal width have approximately Gaussian distribution 
```

```python
# histogram using matplotlib
# individual histograms 
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(221)
hsg = ax.hist(plot_data['sepal_length'], bins=15)
plt.title('sepal_length')
plt.show()
```

```python
# all four histograms in one
def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure()
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=10,ax=ax)
        ax.set_title(var_name+" Distribution")
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()


draw_histograms(plot_data, plot_data.columns, 2, 2)
```

## Using Seaborn

### 4.0 Scatter Plot - Sepal_Length_Width Vs Species.

```python
sns.set_style('whitegrid') # {darkgrid, whitegrid, dark, white, ticks}
fig = sns.FacetGrid(df, hue='species', height=8, aspect=1.5)\
.map(plt.scatter, 'sepal_length', 'sepal_width')\
.add_legend()
```

Observations:<br>
From scatter plots we can identify features that can be used to distinguish between classes 
Here we can see that
1. Setosa can be distinguished from other 2 classes by just using sepal width and length
2. Other two classes cannot be distinguished from the same information as they have a considerable overlap. 
3. This vs comparison is fruitful for distinguishing setosa variant only

4.4 Violinplot - Density of the length and width in the species

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='species',y='petal_length',data=df)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='petal_width',data=df)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='sepal_length',data=df)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='sepal_width',data=df)
plt.show()
```
