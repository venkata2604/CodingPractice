# Activation Functions

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
import seaborn 
import math
import plotly
```

```python
x = np.linspace(-10, 10, 5000)
x
```

# Binary Activation Function

```python
def binary_activation_function(x):
    return 0 if x<0 else 1
```

```python
y = [binary_activation_function(i) for i in x]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x,y,)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title("Binary Activation Function")
plt.show()
```

# Linear Activation Function

# $$ y = mx $$

```python
def linear_activation_function(a,x):
    return a*x
```

```python
y_1 = [linear_activation_function(0.25, i) for i in x]
y_2 = [linear_activation_function(0.5, i) for i in x]
y_3 = [linear_activation_function(1, i) for i in x]
y_4 = [linear_activation_function(2, i) for i in x]
```

```python
plt.figure(num=1, figsize=(10,10))
plt.plot(x, y_1, label='y1')
plt.plot(x, y_2, label='y2')
plt.plot(x, y_3, label='y3')
plt.plot(x, y_4, label='y4')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Activation Function')
plt.legend()
plt.grid()
plt.show()
```

# Sigmoid Activation Function

```python
def sigmoid_activation_function(x):
    return 1/(1+np.exp(-x))
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, [sigmoid_activation_function(i) for i in x])
plt.grid()
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('1/(1+exp(-x)')
plt.show()
```

# tanh Activation Function

# $$ tanh = sinh /cosh $$


# $$tanh(x)=2(2x)1$$
where  is the sigmoid function

```python
def tanh(x):
    return 2*sigmoid_activation_function(2*x) -1
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, [tanh(i) for i in x])
plt.grid()
plt.title('tanh Activation Function')
plt.xlabel('x')
plt.ylabel('1/(1+exp(-x)')
plt.show()
```

# ReLU (Rectified Linear Unit Activation Function

```python
def ReLU(x):
    return np.array([0, x]).max()
```

```python
y = [ReLU(i) for i in x]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y,)
plt.grid()
```

# Softmax

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
x_soft = np.linspace(0, 3.5, 5)
y = softmax(x_soft)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x_soft,y)
plt.grid()
plt.show()
```
