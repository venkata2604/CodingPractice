## Recursion


A recursive function class on itself with to perform the subtasks. 

Basepoint is when the function encounter a subtask that it can perform wihtout calling itself. 

In basecase a function does not recur. 

Recursive case is when the function calls itself to perform a subtask. 

```python
# demonstrating recursion using factorial

#factorial.py
def factorial(n):
    if n==0: return 1
    print(n, end =' ')
    return n*factorial(n-1)


print("\nthe result is ", factorial(6))
```

## Rercursion vs Iteration

Recursion
* Terminates when a base case is reached 
* Each recursive call requires extra space on the stack frame (memory) 
* If we get infinte recursion, the program may run out of memory and result in stack overflow
* Solutions to some problems are easier to formulate recursively

Iteration
* Terminates when a condition is proven to be false
* Each iteration does not require an extra space
* An infinte loop coud loop forever since there is no extra memory being created 
* Iterative solutions to problem may not always be as obvious as a recursive solution

#### Examples of Recursion

* Graph traversals: DFS [Depth First Search] and BFS [Breadth First Search]

* Backtracking Algorithms 

### Towers of Hanoi


https://www.khanacademy.org/computing/computer-science/algorithms/towers-of-hanoi/a/towers-of-hanoi





```python
#Iterative solution
def tower_of_hanoi_iterative(n, source, destination, auxiliary):
    stack = []
    stack.append((n,source, destination, auxiliary))
    while stack:
        disks, source, destination, auxiliary = stack.pop()
        if disks == 1:
            print("Move disk 1 from source", source, 'to destination', destination)
        else:
            stack.append((disks-1, source, auxiliary, destination))
            stack.append((1, source, destination, auxiliary))
            stack.append((disks-1, auxiliary, destination, source))


# Usage
tower_of_hanoi_iterative(3, 'A', 'C', 'B')
```
